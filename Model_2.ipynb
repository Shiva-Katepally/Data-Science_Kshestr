```python
import keras
callback_early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, verbose=0, mode='auto')
```

```python
def build_nn_model(X_train, y_train, X_test, y_test):
    # Define model
    model = keras.models.Sequential()

    model.add(keras.layers.normalization.BatchNormalization(input_shape=tuple([3])))
    model.add(keras.layers.core.Dense(64, activation='relu'))
    model.add(keras.layers.core.Dropout(rate=0.5))
    model.add(keras.layers.normalization.BatchNormalization())
    model.add(keras.layers.core.Dense(32, activation='relu'))
    model.add(keras.layers.core.Dropout(rate=0.5))
    model.add(keras.layers.normalization.BatchNormalization())
    model.add(keras.layers.core.Dense(64, activation='relu'))
    model.add(keras.layers.core.Dropout(rate=0.5))
    model.add(keras.layers.normalization.BatchNormalization())
    model.add(keras.layers.core.Dense(32, activation='relu'))
    model.add(keras.layers.core.Dropout(rate=0.5))
    model.add(keras.layers.normalization.BatchNormalization())
    model.add(keras.layers.core.Dense(32, activation='relu'))
    model.add(keras.layers.core.Dropout(rate=0.5))
    model.add(keras.layers.core.Dense(2,   activation='sigmoid'))
    model.compile(loss="categorical_crossentropy", optimizer="adadelta",metrics=["accuracy"])
    
    y_train_nn = keras.utils.np_utils.to_categorical(y_train)
    y_valid_nn = keras.utils.np_utils.to_categorical(y_test)
    
    model.fit(X_train, y_train_nn, batch_size=1024, epochs=200, validation_data=(X_test, y_valid_nn), verbose=1, callbacks=[callback_early_stopping])
    
    return model
```
